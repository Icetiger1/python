```python
columns = ['name', 'age', 'city']
index = [1,2,3,4]
data = [['grisha', 99, 'Msc'],
        ['katya', 23, 'perm'],
        ['sergey', 45, np.nan],
        ['alex', 104, 'spb']]
df_1 = pd.DataFrame(data = data,
                  columns = columns,
                  index = index)
df_1
```

```python
df_1.info()
```

```python
columns = ['name_2', 'age', 'city_2']
index = [1,2,3,4,5]
data = [['grisha', 99, np.nan],
        ['katya', 23, 'perm'],
        [np.nan, 45, 'spb'],
        ['alex', 104, 'spb'],
        ['irina', 123, 'msc']]
df_2 = pd.DataFrame(data = data,
                  columns = columns,
                  index = index)
df_2
```

```python
df_2.info()
```

```python
#LEFT
df_1.merge(df_2, on = 'age', how ='left')
```

```python
#RIGHT
df_1.merge(df_2, on = 'age', how ='right')
```

```python
#INNER
df_1.merge(df_2, on = 'age', how ='inner')
```

```python
#OUTER
df_1.merge(df_2, on = 'age', how ='outer')
```

```python
#OUTER
df_1[df_1['age'] > 100].merge(df_2, on = 'age', how ='outer')
```
